```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import torch

from analysis.datasets import load_entsoe
from analysis.splits import to_train_validation_test_data
from analysis.TabPFN import train_tabpfn_model, compute_cdf_pdf_interpolators, evaluate
```

```python
entsoe = load_entsoe()
entsoe.head(5)
```

```python
X_validation, y_validation, quantiles_custom, all_quantiles = train_tabpfn_model(entsoe)
```

```python
probabilities = quantiles_custom
y_min = -21
y_max = 0
data_points = 200

crps_cdf_linear_a = []
crps_cdf_pchip_a = []
crps_hybrid_cdf_a = []

nll_pdf_linear_a = []
nll_pdf_linear2_a = []
nll_pdf_pchip_a = []
nll_pdf_hybrid_a = []

for i in range(527, 528):
    quantile_i = all_quantiles[:, i]
    y_i = y_validation[i]
    cdf_linear_a, cdf_pchip_a, hybrid_cdf_a, pdf_linear_a, pdf_linear2_a, pdf_pchip_a, pdf_hybrid_a = evaluate(quantile_i, probabilities, y_i, y_min, y_max)

    crps_cdf_linear_a.append(cdf_linear_a)
    #print(f"{i}. crps linear value: {crps_cdf_linear_a}")

    crps_cdf_pchip_a.append(cdf_pchip_a)
    #rint(f"{i}. crps pchip value: {crps_cdf_pchip_a}")

    crps_hybrid_cdf_a.append(hybrid_cdf_a)
    #print(f"{i}. crps_hybrid_cdf value: {crps_hybrid_cdf_a}")

    nll_pdf_linear_a.append(pdf_linear_a)
    #print(f"{i}. nll_pdf_linear value: {nll_pdf_linear_a}")

    nll_pdf_linear2_a.append(pdf_linear2_a)
    #print(f"{i}. nll_pdf_linear2 value: {nll_pdf_linear2_a}")

    nll_pdf_pchip_a.append(pdf_pchip_a)
    print(f"{i}. nll_pdf_pchip_a value: {nll_pdf_pchip_a}")
    print(y_i)
    print(quantile_i)

    nll_pdf_hybrid_a.append(pdf_hybrid_a)
    print(f"{i}. nll_pdf_hybrid vaue: {nll_pdf_hybrid_a}")

mean_crps_cdf_linear = np.mean(crps_cdf_linear_a)
mean_crps_cdf_pchip_a = np.mean(crps_cdf_pchip_a)
mean_crps_hybrid_cdf_a = np.mean(crps_hybrid_cdf_a)

mean_nll_pdf_linear_a = np.mean(nll_pdf_linear_a)
mean_nll_pdf_linear2_a = np.mean(nll_pdf_linear2_a)
mean_nll_pdf_pchip_a = np.mean(nll_pdf_pchip_a)
mean_nll_pdf_hybrid_a = np.mean(nll_pdf_hybrid_a)

print(f"Mean CRPS for CDF Linear interpolation: {mean_crps_cdf_linear:.4f}")
print(f"Mean CRPS for CDF PCHIP interpolation: {mean_crps_cdf_pchip_a:.4f}")
print(f"Mean CRPS for Hybrid CDF interpolation: {mean_crps_hybrid_cdf_a:.4f}")

print(f"Mean NLL for PDF Linear interpolation: {mean_nll_pdf_linear_a:.4f}")
print(f"Mean NLL for PDF Linear2 interpolation: {mean_nll_pdf_linear2_a:.4f}")
print(f"Mean NLL for PDF PCHIP interpolation: {mean_nll_pdf_pchip_a:.4f}")
print(f"Mean NLL for Hybrid PDF interpolation: {mean_nll_pdf_hybrid_a:.4f}")
```
