```python
import pandas as pd
```

```python
data = {
    'year': [2017, 2018, 2019, 2020, 2021, 2022, 2023],
    'installed_capacity': [17866, 18346, 18711, 19138, 19748, 20414, 21078]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Set 'Year' as the index and convert it to datetime
df['year'] = pd.to_datetime(df['year'], format='%Y')
df.set_index('year', inplace=True)

df
```

```python
df.resample("15min").nearest().plot()
```

```python
df.resample("15min").interpolate(method="linear").plot()
df.resample("15min").interpolate(method="linear")
```
